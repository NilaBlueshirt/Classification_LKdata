```python
import pandas as pd
import numpy as np
```

```python
df_abs = pd.read_csv('order_absolute_reads_AZ.txt', delimiter='\t')
df_n = pd.read_csv('order_normalized_reads_AZ.txt', delimiter='\t')
df_r = pd.read_csv('order_relative_abundance_AZ.txt', delimiter='\t')
df_infant = pd.read_csv('drop_Infants.csv')
```

```python
Inumber = df_infant['Inumber'].to_list()
```

```python
df_abs = df_abs.drop(df_abs[df_abs['sampleID'].isin(Inumber)].index)
df_n = df_n.drop(df_n[df_n['sampleID'].isin(Inumber)].index)
df_r = df_r.drop(df_r[df_r['sampleID'].isin(Inumber)].index)
```

```python
df_abs.loc[(df_abs['CovidStatus']==0) & (df_abs['EverCovid']==0), 'CovidLabel'] = 0
df_abs.loc[(df_abs['CovidStatus']==0) & (df_abs['EverCovid']==1), 'CovidLabel'] = 1
df_abs.loc[(df_abs['CovidStatus']==1) & (df_abs['EverCovid']==1), 'CovidLabel'] = 2
df_abs['CovidLabel'] = df_abs['CovidLabel'].astype(int)
df_abs
```

```python
df_n.loc[(df_n['CovidStatus']==0) & (df_n['EverCovid']==0), 'CovidLabel'] = 0
df_n.loc[(df_n['CovidStatus']==0) & (df_n['EverCovid']==1), 'CovidLabel'] = 1
df_n.loc[(df_n['CovidStatus']==1) & (df_n['EverCovid']==1), 'CovidLabel'] = 2
df_n['CovidLabel'] = df_n['CovidLabel'].astype(int)
df_n
```

```python
df_r.loc[(df_r['CovidStatus']==0) & (df_r['EverCovid']==0), 'CovidLabel'] = 0
df_r.loc[(df_r['CovidStatus']==0) & (df_r['EverCovid']==1), 'CovidLabel'] = 1
df_r.loc[(df_r['CovidStatus']==1) & (df_r['EverCovid']==1), 'CovidLabel'] = 2
df_r['CovidLabel'] = df_r['CovidLabel'].astype(int)
df_r
```

```python
df_abs.to_csv('order_abs_reads_otu_drop.csv', index=False)
df_n.to_csv('order_normalized_reads_otu_drop.csv', index=False)
df_r.to_csv('order_relative_abundance_otu_drop.csv', index=False)
```
